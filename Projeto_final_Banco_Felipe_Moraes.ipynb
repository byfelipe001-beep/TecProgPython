**Contexto:** Você é analista de dados do Banco Fácil. O banco te enviou uma tabela de transações e uma tabela de cadastro de clientes. Seu trabalho é cruzar essas bases e gerar uma análise rápida.

```python
import pandas as pd
import numpy as np

# Base de transações
transacoes = pd.DataFrame({
    'id_cliente': [1, 2, 1, 3, 2, 1, 3, 2],
    'tipo':       ['Crédito', 'Débito', 'Débito', 'Crédito', 'Crédito', 'Débito', 'Débito', 'Crédito'],
    'categoria':  ['Salário', 'Alimentação', 'Transporte', 'Salário', 'Lazer', 'Alimentação', 'Transporte', 'Salário'],
    'valor':      [3200.0, 150.0, None, 4500.0, 200.0, 80.0, None, 2800.0],
    'canal':      ['App', 'Agência', 'App', 'Internet Banking', 'App', 'Agência', 'App', 'Internet Banking']
})

# Base de cadastro de clientes
clientes = pd.DataFrame({
    'id_cliente': [1, 2, 3, 4],
    'nome':       ['Ana', 'Bruno', 'Carlos', 'Diana'],
    'estado':     ['SP', 'RJ', 'MG', 'SP'],
    'perfil':     ['Premium', 'Básico', 'Intermediário', 'Básico']
})
```

### Tarefa 1 — Conheça os dados

Use `shape`, `dtypes`, `head()` e `isna().sum()` nas **duas** tabelas separadamente.

- Quantos clientes existem no cadastro?
- Quantas transações existem?
- Alguma coluna possui valores nulos?

```python
# Exibindo as formas dos DataFrames
print("Shape do df transacoes:", transacoes.shape)
print("Shape do df clientes:", clientes.shape)

# Exibindo as primeiras linhas dos DataFrames
print("\nPrimeiras linhas do df transacoes:")
display(transacoes.head())
print("\nPrimeiras linhas do df clientes:")
display(clientes.head())

# Verificando tipos de dados
print("\nTipos de dados do df transacoes:")
print(transacoes.dtypes)
print("\nTipos de dados do df clientes:")
print(clientes.dtypes)

# Verificando valores ausentes
print("\nValores ausentes no df transacoes:")
print(transacoes.isnull().sum())
print("\nValores ausentes no df clientes:")
print(clientes.isnull().sum())

# Respostas:
print("\nRespostas:")
print("Existem 4 clientes no cadastro")
print("Existem 8 transações registradas")
print("A coluna 'valor' do df transacoes possui 2 valores ausentes")
```

### Tarefa 2 — Limpeza

Preencha os valores nulos da coluna `valor` com a **mediana geral** da coluna.  
Use `fillna()`.

Confirme que não existem mais nulos com `isna().sum()`.

```python
# Preenchendo os valores ausentes na coluna 'valor' com a mediana geral da coluna
transacoes_ajustado = transacoes.fillna(transacoes['valor'].median())
print("\nTransações ajustadas (com valores ausentes preenchidos):")
display(transacoes_ajustado)

# Conferindo o resultado do preenchimento
print("\nValores ausentes após preenchimento:")
print(transacoes_ajustado.isna().sum())
```

### Tarefa 3 — Cálculos com NumPy

Com os valores da coluna `valor` (já limpa), crie um array NumPy e calcule:

- Soma total das transações
- Valor médio
- Maior valor
- Menor valor

```python
# Analisando as transações ajustadas
array = transacoes_ajustado['valor'].values
print(array, "\n")
print("Soma total das transações:", np.sum(array))
print("Valor médio das transações:", np.mean(array))
print("Maior valor das transações:", np.max(array))
print("Menor valor das transações:", np.min(array))
```
