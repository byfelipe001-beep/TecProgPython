## 📦 1. Criação dos DataFrames

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Clientes
df_clientes = pd.DataFrame({
    'id_cliente': np.arange(1, 101),
    'nome': ['Cliente_' + str(i) for i in range(1, 101)],
    'idade': np.random.choice([25, 30, 35, 40, np.nan], 100),
    'cidade': np.random.choice(['RJ', 'SP', 'MG', None], 100),
    'status_conta': np.random.choice(['Ativa', 'Inativa', None], 100),
    'score_credito': np.random.choice([300, 500, 700, 900, np.nan], 100),
    'renda_mensal': np.random.choice([2000, 4000, 6000, 8000, np.nan], 100),
    'segmento': np.random.choice(['Varejo', 'Alta Renda', None], 100)
})

# Inserindo duplicados
df_clientes = pd.concat([df_clientes, df_clientes.iloc[:5]])

# Transações
df_transacoes = pd.DataFrame({
    'id_cliente': np.random.randint(1, 101, 200),
    'valor_transacao': np.random.choice([100, 500, 1000, 2000, np.nan], 200),
    'tipo': np.random.choice(['Crédito', 'Débito'], 200),
    'mes': np.random.choice(['Jan', 'Fev', 'Mar'], 200)
})

df_clientes.head()
```

### 🧹 PARTE 3 — Limpeza de dados

**Pergunta 5**  
Remova duplicados  

**Pergunta 6**  
Preencha valores faltantes:
- idade → mediana  
- renda → média  
- cidade → "Não Informado"  
- status → ffill  


```python
df_clientes_limpo = df_clientes.drop_duplicates()
df_clientes_limpo['idade'] = df_clientes_limpo['idade'].fillna(df_clientes_limpo['idade'].median())
df_clientes_limpo['renda_mensal'] = df_clientes_limpo['renda_mensal'].fillna(df_clientes_limpo['renda_mensal'].mean())
df_clientes_limpo['cidade'] = df_clientes_limpo['cidade'].fillna('Não Informado')
df_clientes_limpo['status_conta'] = df_clientes_limpo['status_conta'].ffill().bfill()
display(df_clientes_limpo.head())
df_clientes_limpo.isna().sum()  
```

### 🔁 PARTE 4 — Regras de negócio

**Pergunta 7**  
Crie a coluna `perfil_risco`:
- "Alto Risco" → score < 500  
- "Médio" → 500–700  
- "Baixo Risco" → > 700  


```python
df_clientes_limpo['perfil_risco'] = np.select(
    [
        df_clientes_limpo['score_credito'] < 500,
        df_clientes_limpo['score_credito'].between(500, 700),
        df_clientes_limpo['score_credito'] > 700
    ],
    ['Alto Risco', 'Médio', 'Baixo Risco'],
    default='Desconhecido'
)
```
