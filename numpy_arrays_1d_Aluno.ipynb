### 📖 O que é NumPy e por que usá-lo?

**NumPy** (Numerical Python) é a biblioteca fundamental para computação numérica em Python. Ela introduz o conceito de **array** — uma estrutura de dados otimizada para armazenar e manipular grandes coleções de números de forma eficiente.

Em Data Science, quase tudo envolve números: preços, temperaturas, idades, distâncias, porcentagens. O NumPy é o alicerce sobre o qual bibliotecas famosas como Pandas, Scikit-learn e TensorFlow são construídas.

> 🧱 **Analogia da construção civil:** Imagine que o Python é uma casa e o NumPy é a fundação de concreto armado. Você pode construir uma casa em cima de areia (listas Python), mas ela vai ser lenta e instável. Com o NumPy como fundação, tudo fica mais sólido e rápido.

**Por que o NumPy é mais rápido do que listas Python?**

Uma lista Python é como uma gaveta de escritório onde cada compartimento pode guardar qualquer coisa: um papel, uma régua, um grampeador. Essa flexibilidade tem um custo: toda vez que você quer pegar algo, o Python precisa verificar o que está ali.

Um array NumPy é como uma caixa de ovos: todos os compartimentos têm o mesmo tamanho e guardam exatamente o mesmo tipo de coisa. O Python sabe exatamente onde cada ovo está, sem precisar verificar nada. Resultado? Operações até **100x mais rápidas** em cálculos numéricos.

### Criação de Arrays 1D

Um array 1D (unidimensional) é simplesmente uma sequência de números — como uma linha de uma planilha. Existem cinco formas principais de criar um array 1D:

#### `np.array()` — A partir de uma lista existente

É o ponto de entrada mais comum. Você pega dados que já tem (em uma lista Python) e transforma em array NumPy.

> 🍱 **Analogia da marmita:** Você tem os ingredientes separados (lista Python) e coloca tudo na marmita organizada (array NumPy). Os ingredientes são os mesmos, mas agora estão prontos para o trabalho pesado.

#### `np.arange()` — Sequências com passo definido

Funciona como o `range()` do Python, mas retorna um array NumPy. Ideal para criar sequências numéricas com início, fim e passo.

#### `np.linspace()` — Sequências com número fixo de pontos

Em vez de definir o passo, você define *quantos pontos* quer entre dois valores. O NumPy divide o intervalo igualmente. Muito usado para gráficos e interpolações.

> 📏 **Analogia da régua:** `arange` é como marcar a cada 1 cm. `linspace` é como dividir a régua em exatamente 10 partes iguais, não importa o tamanho dela.

#### `np.zeros()` e `np.ones()` — Arrays de inicialização

Criam arrays preenchidos com zeros ou uns. Muito usados para inicializar variáveis antes de preencher com dados reais, como deixar uma planilha em branco antes de digitar os valores.

---

### Atributos dos Arrays

Depois de criar um array, você pode inspecioná-lo com quatro atributos essenciais:

| Atributo | O que informa | Exemplo |
|----------|---------------|---------|
| `.shape` | Dimensões do array (como tupla) | `(5,)` para 1D com 5 elementos |
| `.dtype` | Tipo dos dados armazenados | `int64`, `float64` |
| `.ndim` | Número de dimensões | `1` para arrays 1D |
| `.size` | Total de elementos | `5` |


> 🗂️ **Analogia do arquivo morto:** `.shape` é o número de gavetas e a profundidade de cada uma. `.dtype` é o tipo de documento guardado. `.ndim` é quantas dimensões tem o armário (prateleiras, gavetas, caixas). `.size` é o total de documentos.

### Passo 3: Criando arrays com `np.arange()`

````python
# np.arange(início, fim, passo) — o valor de "fim" NÃO é incluído
dias_do_mes = np.arange(1, 32, 1)  # Dias de 1 a 31
print(dias_do_mes)

# Passo diferente: semanas do ano (a cada 7 dias)
semanas = np.arange(0, 365, 7)
print(semanas)

# Funciona com floats também
decimais = np.arange(0.0, 1.1, 0.2)  # De 0.0 a 1.0 com passo 0.2
print(decimais)
````

### Passo 4: Criando arrays com `np.linspace()`

````python
# np.linspace(início, fim, quantidade) — divide o intervalo em partes IGUAIS
# O valor de "fim" É incluído (diferente do arange!)

# 5 pontos igualmente espaçados entre 0 e 1
pontos = np.linspace(0, 1, 5)
print(pontos)

# 10 temperaturas entre 20°C e 40°C (para um gráfico)
temperaturas = np.linspace(20, 40, 10)
print(temperaturas)
````

### Passo 5: Criando arrays com `np.zeros()`,  `np.ones()` e `np.full()`

````python
# np.zeros(n) — cria um array com n zeros (útil para inicializar variáveis)
zeros = np.zeros(5)
print("Zeros:", zeros)

# np.ones(n) — cria um array com n uns
uns = np.ones(6)
print("Uns:", uns)

#np.full(n, valor) — cria um array com n elementos, todos iguais a "valor"
full = np.full(5, 2)
print("Full:", full)

# Especificando o tipo de dado (inteiro em vez de float)
zeros_int = np.zeros(5, dtype=int)
print("Zeros inteiros:", zeros_int)
````

### Indexação e Fatiamento (Slicing)

Acessar elementos de um array funciona igual às listas Python, com uma vantagem extra: indexação negativa e fatiamento poderoso.

- **Índice positivo:** começa do 0 (primeiro elemento)
- **Índice negativo:** começa do -1 (último elemento)
- **Slicing:** `array[início:fim:passo]` — o fim é **exclusivo**

> 🚌 **Analogia do ônibus:** Imagine um ônibus com assentos numerados de 0 a 9. `array[0]` é o banco da janela da frente. `array[-1]` é o último banco. `array[2:5]` são os bancos 2, 3 e 4 (o 5 fica de fora, como o passageiro que perdeu o ônibus).

---

### Passo 7: Indexação e fatiamento

````python
notas = np.array([7.5, 8.0, 6.5, 9.2, 5.8, 10.0, 4.3])
#  índice:          0    1    2    3    4     5     6
# índice negativo: -7   -6   -5   -4   -3    -2    -1

# Indexação simples (positiva)
print("Primeira nota :", notas[0])   # Índice 0 = primeiro elemento
print("Quarta nota   :", notas[3])   # Índice 3 = quarto elemento

# Indexação negativa (de trás para frente)
print("Última nota   :", notas[-1])  # -1 = último elemento
print("Penúltima nota:", notas[-2])  # -2 = penúltimo elemento

# Fatiamento: array[início:fim]  (fim é EXCLUSIVO)
print("Notas 1 a 3   :", notas[1:4])   # Índices 1, 2, 3
print("Primeiras 3   :", notas[:3])    # Do início até o índice 2
print("Últimas 2     :", notas[-2:])   # Do penúltimo até o fim

# Fatiamento com passo: array[início:fim:passo]
print("Intercaladas  :", notas[::2])   # Pega de 2 em 2 (índices 0, 2, 4, 6)
print("Invertido     :", notas[::-1])  # Inverte o array
````

### Operações Matemáticas Element-wise e Broadcasting

**Element-wise (elemento a elemento):** Quando você soma dois arrays de mesmo tamanho, o NumPy soma os elementos que estão na mesma posição. É como somar duas colunas de uma planilha linha por linha.

**Broadcasting básico:** Quando você soma um array com um número simples (escalar), o NumPy aplica a operação em **todos os elementos** de uma vez. É como dar um aumento de 10% para todos os funcionários ao mesmo tempo, em vez de calcular um por um.

> 💰 **Analogia do salário:** Você tem uma lista de salários. Dar um aumento de R$ 500 para todos é broadcasting: `salarios + 500`. Calcular o salário líquido de cada um (salário - desconto) é element-wise: `salarios - descontos`.

---

### Passo 8: Operações element-wise e broadcasting

````python
# Salários de 5 funcionários (em R$)
salarios = np.array([3500, 4200, 2800, 5100, 3900])

# --- Broadcasting: operação com um escalar (número único) ---
# Dar aumento de R$ 500 para todos
salarios_reajustados = salarios + 500
print("Após aumento de R$500 :", salarios_reajustados)

# Calcular INSS de 11% de cada salário
inss = salarios * 0.11
print("INSS (11%)            :", inss)

# --- Element-wise: operação entre dois arrays de mesmo tamanho ---
# Bônus individual de cada funcionário
bonus = np.array([200, 500, 150, 800, 300])

# Salário total = salário + bônus (elemento a elemento)
total = salarios + bonus
print("Salário total         :", total)

# Comparação element-wise: quem ganha mais de R$ 4000?
acima_4000 = salarios > 4000
print("Ganha acima de R$4000 :", acima_4000)  # Retorna array de True/False

# Média salarial
media = np.mean(salarios)
print("Média salarial        :", media)

# Desvio padrão salarial
desvio = np.std(salarios)
print("Desvio padrão         :", desvio)
````

### Funções Matemáticas e Estatísticas

O NumPy oferece funções prontas para os cálculos mais comuns em análise de dados:

| Função | O que calcula |
|--------|---------------|
| `np.sum()` | Soma de todos os elementos |
| `np.mean()` | Média aritmética |

| `np.min()` | Menor valor |
| `np.max()` | Maior valor |

> 📊 **Analogia do boletim escolar:** É como pegar as notas do semestre e calcular de uma vez: a soma total, a média, o desvio, a menor e a maior nota — sem precisar de calculadora.

### Passo 9: Funções matemáticas e estatísticas


````python
# Faturamento diário de uma loja (R$) na última semana
faturamento = np.array([4523.10, 3891.45, 5102.80, 6230.00, 2987.50, 7540.20, 5312.75])

print(f"Total da semana   : R$ {np.sum(faturamento):,.2f}")    # Soma
print(f"Média diária      : R$ {np.mean(faturamento):,.2f}")   # Média
print(f"Desvio padrão     : R$ {np.std(faturamento):,.2f}")    # Dispersão dos valores
print(f"Pior dia          : R$ {np.min(faturamento):,.2f}")    # Menor valor
print(f"Melhor dia        : R$ {np.max(faturamento):,.2f}")    # Maior valor

# Identificar índice do melhor e pior dia
print(f"Índice do pior dia  : {np.argmin(faturamento)}")       # np.argmin: posição do mínimo
print(f"Índice do melhor dia: {np.argmax(faturamento)}")       # np.argmax: posição do máximo

# grafico de barras do faturamento diário com linhas de média e desvio padrão  
import matplotlib.pyplot as plt
dias = np.arange(1, 8)  # Dias da semana (1 a 7)
plt.bar(dias, faturamento, color='skyblue')
plt.axhline(np.mean(faturamento), color='red', linestyle='--', label='Média')
plt.axhline(np.mean(faturamento) + np.std(faturamento), color='green', linestyle='--', label='Média + Desvio')
plt.axhline(np.mean(faturamento) - np.std(faturamento), color='orange', linestyle='--', label='Média - Desvio')
plt.xlabel('Dia da Semana')
plt.ylabel('Faturamento (R$)')
plt.title('Faturamento Diário da Loja')
plt.legend()
plt.xticks(dias)  # Define os rótulos do eixo x como os dias
plt.show()
````

### Passo 10: Comparativo NumPy vs Lista Python (velocidade)


````python
import time  # Biblioteca para medir tempo de execução

# Criando 100 milhões de números para o teste (aumentado para melhor medição)
tamanho = 100_000_000

lista_python = list(range(tamanho))           # Lista Python nativa
array_numpy  = np.arange(tamanho)             # Array NumPy

# --- Teste com lista Python ---
inicio = time.time()
resultado_lista = [x * 2 for x in lista_python]  # Multiplicar por 2 usando list comprehension
fim = time.time()
tempo_lista = fim - inicio

# --- Teste com NumPy ---
inicio = time.time()
resultado_numpy = array_numpy * 2                  # Multiplicar por 2 com NumPy (broadcasting)
fim = time.time()
tempo_numpy = fim - inicio

# Evitar divisão por zero caso tempo_numpy seja 0
if tempo_numpy > 0:
	speedup = tempo_lista / tempo_numpy
	print(f"Tempo com lista Python : {tempo_lista:.4f} segundos")
	print(f"Tempo com NumPy        : {tempo_numpy:.4f} segundos")
	print(f"NumPy foi {speedup:.1f}x mais rápido")
else:
	print(f"Tempo com lista Python : {tempo_lista:.4f} segundos")
	print(f"Tempo com NumPy        : {tempo_numpy:.4f} segundos")
	print("NumPy é tão rápido que o tempo é imeasurável com esta precisão!")
````

````python
equipe_a = np.array([42000, 38500, 51000, 47300, 55800, 49200])
equipe_b = np.array([39000, 41200, 48700, 52000, 44600, 53100])

print("Faturamento total da Equipe A:", np.sum(equipe_a))
print("Faturamento total da Equipe B:", np.sum(equipe_b))

if np.sum(equipe_a) > np.sum(equipe_b):
    print("A Equipe A tem maior faturamento total."  )
else:
    print("A Equipe B tem maior faturamento total.")

diferenca = equipe_a - equipe_b
print("Diferença de faturamento (A - B):", diferenca)

faturou_mais = diferenca > 0
# se a equipe A faturou mais, imprime em qual mês isso aconteceu usando len(diferenca) para contar os meses
if faturou_mais.any() == True:
    print("A Equipe A faturou mais que a B em:", np.sum(faturou_mais), "meses.")
    meses = np.where(faturou_mais)[0] + 1  # Índices 0-based para 1-based
    print("Meses em que A faturou mais:", meses)


print('Média Equipe A:', np.round(np.mean(equipe_a), 2))  # Média da equipe A
print('Média Equipe B:', np.round(np.mean(equipe_b), 2))  # Média da equipe B
print('Desvio Padrão Equipe A:', np.round(np.std(equipe_a), 2))   # Desvio padrão da equipe A
print('Desvio Padrão Equipe B:', np.round(np.std(equipe_b), 2))   # Desvio padrão da equipe B

if np.std(equipe_a) < np.std(equipe_b):
    print("A Equipe A tem faturamento mais consistente (menor desvio padrão).")
else:    
    print("A Equipe B tem faturamento mais consistente (menor desvio padrão).")

bonus = np.array([0.08])
bonus_equipe_a = equipe_a * bonus
bonus_equipe_b = equipe_b * bonus

print('Bônus Equipe A:', np.round(bonus_equipe_a, 2))  # Bônus da equipe A
print('Bônus Equipe B:', np.round(bonus_equipe_b, 2))
print('Total de bonus equipe A:', np.round(np.sum(bonus_equipe_a), 2))  # Total de bônus da equipe A
print('Total de bonus equipe B:', np.round(np.sum(bonus_equipe_b), 2))  # Total de bônus da equipe B
````

````python
equipe_a = np.array([42000, 38500, 51000, 47300, 55800, 49200])
equipe_b = np.array([39000, 41200, 48700, 52000, 44600, 53100])

# 1. Total de cada equipe
total_a = np.sum(equipe_a)
total_b = np.sum(equipe_b)
print(f"Total Equipe A: R$ {total_a:,}")
print(f"Total Equipe B: R$ {total_b:,}")
vencedor = "A" if total_a > total_b else "B"
print(f"Maior faturamento total: Equipe {vencedor}")

print()

# 2. Diferença mês a mês (element-wise)
diferenca = equipe_a - equipe_b
print("Diferença (A - B) por mês:", diferenca)
meses_a_venceu = diferenca > 0   # Array de True/False
print("Meses em que A superou B:", meses_a_venceu)
# np.where retorna os índices onde a condição é True
print("Índices dos meses (A venceu):", np.where(meses_a_venceu)[0])

print()

# 3. Média e desvio padrão
print(f"Média A : R$ {np.mean(equipe_a):,.2f} | Desvio: R$ {np.std(equipe_a):,.2f}")
print(f"Média B : R$ {np.mean(equipe_b):,.2f} | Desvio: R$ {np.std(equipe_b):,.2f}")
mais_consistente = "A" if np.std(equipe_a) < np.std(equipe_b) else "B"
print(f"Equipe mais consistente: Equipe {mais_consistente}")

print()

# 4. Bônus de 8% — broadcasting (multiplica todo array por 0.08)
bonus_a = equipe_a * 0.08
bonus_b = equipe_b * 0.08
print(f"Total de bônus Equipe A: R$ {np.sum(bonus_a):,.2f}")
print(f"Total de bônus Equipe B: R$ {np.sum(bonus_b):,.2f}")

````
